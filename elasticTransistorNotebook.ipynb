# Model for the electro-mechanical behavior of elastic organic transistors

Reynolds, Veronica G., Saejin Oh, Renxuan Xie, and Michael L. Chabinyc. Journal of Materials Chemistry C 8, no. 27 (2020): 9276-9285.

https://doi.org/10.1039/D0TC01181A

The stretchableCircuit module allows you to create virtual stretchable transistors and inverters and plot their strain-dependent behavior.

```python
import stretchableCircuit
```

## Define transistors

```python
OECTntype = stretchableCircuit.transistor(flavor = 'OECT',
                                           Ttype = 'n', 
                                           W = 1, 
                                           L = 1, 
                                           d = 1E-7, 
                                           C = 40000000, 
                                           mu = 1E-4, 
                                           V_T = 0.5,
                                           V_DD = 3,
                                           V_resolution = 100,
                                           deformMode = 'uniaxial-L',
                                           er = [0.5, 0.75, 1, 1.5, 2])

OECTntype.calculateStrainDependence()
OECTntype.calculateI_SD()
OECTntype.calculateRelativeI_SD()
```

```python
OECTptype = stretchableCircuit.transistor(flavor = 'OECT',
                                           Ttype = 'p', 
                                           W = 1, 
                                           L = 1, 
                                           d = 1E-7, 
                                           C = 40000000, 
                                           mu = 1E-4, 
                                           V_T = -0.5,
                                           V_DD = 3,
                                           V_resolution = 100,
                                           deformMode = 'uniaxial-W',
                                           er = [0.5, 0.75, 1, 1.5, 2])

OECTptype.calculateStrainDependence()
OECTptype.calculateI_SD()
OECTptype.calculateRelativeI_SD()
```

## Define inverter

```python
OECTinverter = stretchableCircuit.inverter(OECTntype, OECTptype)
OECTinverter.buildVTC()
```

## Generate plots

### Plot current-voltage curves at varying extension ratios

```python
OECTntype.plotIVvsDeformation([0.5, 0.75, 1, 1.5, 2])
```

```python
OECTptype.plotIVvsDeformation([0.5, 0.75, 1, 1.5, 2])
```

### Plot the current in the saturation regime relative to the undeformed state vs. extension ratio

```python
OECTntype.plotRelativeI_SD()
```

```python
OECTptype.plotRelativeI_SD()
```

### Plot the load curves of the n-type (solid) and p-type (dashed) transistors in the inverter

```python
OECTinverter.plotLoadCurves([0, 0.5, 1, 1.5, 2, 2.5, 3], 1)
```

```python
OECTinverter.plotLoadCurves_alternative([0, 0.5, 1, 1.5, 2, 2.5, 3], 1)
```

### Plot the voltage transfer curve of the inverter at varying extension ratios

```python
OECTinverter.plotVTC([0.5, 0.75, 1, 1.5, 2])
```

```python
OECTinverter.plotVTCeye([0.5, 0.75, 1, 1.5, 2])
```
